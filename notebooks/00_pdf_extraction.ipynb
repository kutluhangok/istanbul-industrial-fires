## Goal
Extract incident-level KMO appendix data for 2017-2023 from PDFs and include the manually provided 2024 Excel table. The important validation point is that the final raw data should now include 2024 with 720 rows.

```python
from pathlib import Path
import sys, json
import pandas as pd
import numpy as np
sys.path.append(str(Path.cwd().parent))

raw = pd.read_excel('../data/raw/kmo_incidents_raw.xlsx')
clean = pd.read_excel('../data/processed/kmo_incidents_clean.xlsx')
log = pd.read_csv('../data/raw/extraction_log.csv')
summary = raw.groupby('source_year').size().rename('rows').reset_index()
summary
```

```python
log
```

```python
validation = {
    'raw_rows': len(raw),
    'clean_rows': len(clean),
    'years': sorted(clean['year'].dropna().astype(int).unique().tolist()),
    '2024_rows': int((clean['year'] == 2024).sum()),
    '2024_fire_explosion': clean.loc[clean['year'] == 2024, 'olay_turu'].value_counts().to_dict(),
    'istanbul_rows_total': int(clean['is_istanbul'].sum()),
}
validation
```

## Interpretation
The dataset now covers 2017-2024 and contains 3,852 cleaned incident rows. The 2024 data was not OCR-derived anymore; it was loaded from the manually provided Excel file and contributes 720 rows. This resolves the earlier 2024 limitation and makes the final panel more complete.
